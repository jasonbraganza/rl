Create an address-book program.  The data will be stored as a list of tuples, with
each tuple containing the following four pieces of information:

    - first name
    - last name
    - phone number
    - email address

The program, when run, will ask the user to enter a one-letter
command.  Each command will do the following:

- q: quit from the program

- a: add a new person. The program will ask four questions, use the
     answers to create a tuple, and then add the tuple to a list.

- l: list all people in the address book

- f: find a person. Ask for a search string, and then display all
    records with that string anywhere in the first name, last name, or
    email address

- w: write the contents of the address book to disk. Use a simple
    tab-delimited CSV format, in which each person will be on one
    line.  Any previously saved info should be overwritten.

- r: read the contents of the address book from disk, replacing the
    existing list of tuples.

Any other one-letter command will cause an error.

```python

letter = None

addbk = []
filename = 'addbk.xsv'

while True:
    letter = input('Enter a command. Do you want to (q)uit, (a)dd an entry, (l)ist people, (f)ind a person, (w)rite out an address book or (r)ead in your address book:  ' ).strip().lower()
    if letter == 'q':
            break
    
    elif letter == 'a':
        first_name = input('Enter first name: ').strip()
        last_name = input('Enter last name: ').strip()
        phone = input('Enter phone: ').strip()
        email = input('Enter email: ').strip()   
        
        new_person = (first_name, last_name, phone, email)
        addbk.append(new_person)
    
    elif letter == 'w':
        with open(filename, 'w') as outfile:
            for one_person in addbk:
                outfile.write('\t'.join(one_person) + '\n')
                
    elif letter == 'r':
        addbk = []
        for one_person in open(filename):
            first_name, last_name, phone, email = one_person.strip().split('\t')
            addbk.append((first_name, last_name, phone, email))
            
    elif letter == 'l':
        for one_person in addbk:
            print(one_person)
    
    elif letter == 'f':
        search_something = input('What are you looking for? ').strip()
        for one_person in addbk:
            first, last, phone, email = one_person
            if (search_something in first or 
                search_something in last or
                search_something in email):
                print(one_person)
    else:
        print('Not a valid command')
        continue
```

Create an address-book program.  The data will be stored as a DICT OF
DICTS, with each dict containing the following four pieces of
information:

    - first name
    - last name
    - phone number
    - email address -- assume that e-mail address is unique

The program, when run, will ask the user to enter a one-letter
command.  Each command will do the following:

- q: quit from the program

- a: add a new person. The program will ask four questions, use the
     answers to create a dict, and then add the dict to a list.

- l: list all people in the address book

- e: Enter an e-mail address, and get the person's info

- f: find a person. Ask for a search string, and then display all
    records with that string anywhere in the first name, last name, or
    email address

- w: write the contents of the address book to disk. Use a simple
    tab-delimited CSV format, in which each person will be on one
    line.  Any previously saved info should be overwritten.

- r: read the contents of the address book from disk, replacing the
    existing list of dict.


Any other one-letter command will cause an error.

```python
letter = None

addbk = {}
filename = 'addbk.xsv'
fields = ['first_name', 'last_name', 'phone',]


while True:

    letter = input('Enter a command. Do you want to (q)uit, (a)dd an entry, (l)ist people, (f)ind a person, (w)rite out an address book or (r)ead in your address book:  ' ).strip().lower()
    if letter == 'q':
            break
    
    elif letter == 'a':
        new_person = {}
        for each_field in fields:
            new_person[each_field] = input(f'Enter {each_field}: ').strip()
        email = input(f'Enter email: ').strip()
        addbk[email]=new_person
    
    elif letter == 'w':
        with open(filename, 'w') as outfile:
            for email, each_person in addbk.items():
                values = list(each_person.values())
                values.append(email)
                outfile.write('\t'.join(values) + '\n')

    elif letter == 'r':
        for each_line in open(filename):
            first_name, last_name, phone, email = each_line.strip().split('\t') 
            addbk[email] = {'first_name': first_name, 'last_name': last_name, 'phone': phone}

    elif letter == 'l':
        for email, each_person in addbk.items():
            first_name, last_name, phone = each_person.values()
            print(f"{last_name}, {first_name}'\t'{phone}, {values}")
    
    elif letter == 'f':
        search_something = input('What are you looking for? ').strip()
        for email, each_person in addbk.items():
            first_name, last_name, phone = each_person.values()
            if (search_something in first_name or 
                search_something in last_name or
                search_something in email):
                print(f"{last_name}, {first_name}'\t'{phone}, {values}")
    
    elif letter == 'e':
        email = input('What email you looking for? ').strip()
        if email in addbk:
            each_person = addbk[email]
            first_name, last_name, phone = each_person.values()
            print(f"{last_name}, {first_name}'\t'{phone}, {values}")
        else:
            print(f"{email}, not in system")
        
    else:
        print('Not a valid command')
        continue
```

Write a function ("vowel_words") that takes a filename as an argument.
It returns a dict with five keys (a, e, i, o, and u), whose values are
lists, containing the words in which those vowels appear. A word may
well appear in more than one value list.

```python
def vowel_words(some_filename):
    vowel_dict = {'a':[], 'e':[], 'i':[], 'o':[],'u':[]}
    for each_line in open(some_filename):
        for each_word in each_line.split():
            for each_vowel, each_list in vowel_dict.items():
                if each_vowel in each_word:
                    vowel_dict[each_vowel].append(each_word)
    return vowel_dict
```
