````python
class Animal(object):
    def __init__(self, colour):
        self.colour = colour
        self.legs = 4
        self.species = self.__class__.__name__
    def __repr__(self):
        return f'{self.colour} {self.species}, {self.legs} legs'
class Wolf(Animal):
    def __init__(self, colour):
        super().__init__(colour)


class Sheep(Animal):
    def __init__(self, colour):
        super().__init__(colour)


class Snake(Animal):
    def __init__(self, colour):
        super().__init__(colour)
        self.legs = 0

class Parrot(Animal):
    def __init__(self, colour):
        super().__init__(colour)
        self.legs = 2
    
class Cage(object):
    def __init__(self, id):
        self.id = id
        self.animals = []
    def add_animals(self, *new_animals):
        self.animals += new_animals
    def __repr__(self):
        output = f'Cage {self.id} \n'
        output += '\n'.join([f'\t {index}. {one_animal}' for index, one_animal in enumerate(self.animals, 1)])
        return output
    def animals_by_colour(self, colour):
        return '\n'.join([str(one_animal) for one_animal in self.animals if one_animal.colour == colour])

class Zoo(object):
    def __init__(self):
        self.cages = []
    def add_cages(self, *new_cages):
        self.cages += new_cages
    def __repr__(self):
        return '\n'.join([str(one_cage) for one_cage in self.cages])
    def animals_by_colour(self, colour):
        return '\n'.join([one_cage.animals_by_colour(colour) for one_cage in self.cages])
    def animals_by_col(self, colour):
        return '\n'.join([str(one_animal)
                             for one_cage in self.cages
                             for one_animal in one_cage.animals
                             if one_animal.colour == colour])
        
````

````python
wolf = Wolf('Black')
sheep1 = Sheep('White')
sheep2 = Sheep('Black')
parrot = Parrot('Red')
snake = Snake('Silver')

c1 = Cage(1)
c1.add_animals(wolf, sheep1, sheep2)

c2 = Cage(2)
c2.add_animals(snake, parrot)

z = Zoo()
z.add_cages(c1, c2)

````
